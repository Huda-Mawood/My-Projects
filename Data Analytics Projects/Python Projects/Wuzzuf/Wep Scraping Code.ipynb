```python
'''
Here I will use the pandas to be able to create a data frame, which is for all my data
'''
import pandas as pd
```

```python
df=pd.DataFrame()
df
```

```python
df=pd.DataFrame(columns=["JobName","CompanyName","LocationJob","JobType","Type_of_Experience","Years_of_Experience","Skills"])
df
```

```python
'''
All I did in the beginning was for one page of the site, but I want to get the data for the entire site,
 so I will create a loop to get the data for all the pages of the site.
'''
```

```python
for i in range(860):
  Request=requests.get(f"https://wuzzuf.net/search/jobs/?a=hpb&q=&start={i}")
  content=Request.content
  soup = BeautifulSoup(content, 'html.parser')
  All_Jobs = soup.find_all("div", class_ = "css-1gatmva e1v1l3u10")
  for Job in All_Jobs:
    JobName=Job.find_all("h2",class_="css-m604qf")[0].text
    CompanyName=Job.find_all("div", class_="css-d7j1kk")[0].text.split("-")[0]
    LocationJob=Job.find_all("span",class_="css-5wys0k")[0].text
    JobType=Job.find_all("span",class_="css-1ve4b75 eoyjyou0")[0].text
    Type_of_Experience=Job.find_all("div",class_="css-1lh32fc")[0].next_sibling.text.split(" · ")[0]
    Years_of_Experience =Job.find_all("div",class_="css-1lh32fc")[0].next_sibling.text.split(" · ")[1]
    Skills =Job.find_all("div",class_="css-1lh32fc")[0].next_sibling.text.split(" · ")[2:]
    df.loc[len(df.index)]=[JobName,CompanyName,LocationJob,JobType,Type_of_Experience,Years_of_Experience,Skills]
  if(i%80==0):
    print("hohod")

```
