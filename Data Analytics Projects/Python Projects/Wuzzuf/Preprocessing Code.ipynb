```python
'''
This is the data you obtained from the site wuzzuf
'''
import pandas as pd
df=pd.read_csv("wuzzuf.csv")
df
```

Delete any column that is not important

```python
'''
First, I will see if I have any null value or not, and I will try to modify the data so that there is no null value.
'''
df=df.drop(columns="Unnamed: 0")
df
```

```python
def clean_JobName(JobName):
  cleaned_JobName=JobName.split("-")[0].strip().split(",")[0].strip().split("/")[0].strip().split("(")[0].strip()
  return cleaned_JobName
df["JobName"]=df["JobName"].apply(clean_JobName)
df["JobName"]
```

```python
def clean_LocationJob(LocationJob):
  cleaned_LocationJob	=LocationJob.split(",")[0].strip()
  return cleaned_LocationJob
df["LocationJob"]=df["LocationJob"].apply(clean_LocationJob	)

```

```python
def Clean_JobType(JobType):
  cleaned_JobType=JobType.split(" / ")[0].strip()
  return cleaned_JobType
df["JobType"]=df["JobType"].apply(Clean_JobType)
```
