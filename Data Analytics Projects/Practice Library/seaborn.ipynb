```python

data=sns.FacetGrid(df,row="smoker",col="time")
data.map(plt.hist,"total_bill",bins=13,color="green")
```

```python
graph=sns.PairGrid(df)
graph=graph.map_upper(sns.scatterplot)
graph=graph.map_lower(sns.kdeplot)
graph=graph.map_diag(sns.histplot)
```

```python
graph=sns.PairGrid(df,hue="day")
graph=graph.map_diag(sns.histplot)
graph=graph.map_offdiag(sns.scatterplot)
graph=graph.add_legend()
```
