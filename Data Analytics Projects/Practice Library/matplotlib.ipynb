```python
#change the size of the markers 
x=[1,7,12,14,6,8,3,9]
plt.plot(x,'o-',ms=20)# ms mean marker size 
```

```python
#Marker Color
#markeredgecolor
x=[1,7,12,14,6,8,3,9]
plt.plot(x,'o-',ms=20,mec='r')# mec mean marker edge color
```

```python
#markerfacecolor
x=[1,7,12,14,6,8,3,9]
plt.plot(x,'*-',ms=20,mfc='m')#mfs mean marker face color
```

```python
#to use both the mec and mfc arguments to color the entire marker
x=[1,7,12,14,6,8,3,9]
plt.plot(x,'*-',ms=20,mec='m',mfc='m')
```

```python
#You can use Hexadecimal color values
x=[1,7,12,14,6,8,3,9]
plt.plot(x,  '*-', ms = 20, mec = '#4CAF50', mfc = '#4CAF50')
```

```python
#Create Labels for a Plot
#Create a Title for a Plot
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("my graph")
```

```python
#Change Font Properties for Title and Labels
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
font1 = {'color':'r','size':20}
font2 = {'color':'k','size':15}
plt.xlabel("X",fontdict=font1)
plt.ylabel("Y",fontdict=font2)
plt.title("my graph",fontdict=font1)
plt.plot(x,y)
```

```python
#Matplotlib Pie Charts
#A simple pie chart
x=np.array([30,25,40,35])
plt.pie(x)
```

```python
#Matplotlib Pie Charts
#Labels
x=np.array([30,25,40,35])
mylabels=np.array(['task','home','school','sleep'])
plt.pie(x,labels=mylabels)
```

```python
#Matplotlib Pie Charts
#Start Angle
x=np.array([30,25,40,35])
mylabels=np.array(['task','home','school','sleep'])
plt.pie(x,labels=mylabels,startangle=90)
```

```python
#Matplotlib Pie Charts
#Explode
x=np.array([30,25,40,35])
mylabels=np.array(['task','home','school','sleep'])
myexplode=np.array([.3,0,0,0])
plt.pie(x,labels=mylabels,explode=myexplode)
```

```python
#Matplotlib Pie Charts
#Shadow
x=np.array([30,25,40,35])
mylabels=np.array(['task','home','school','sleep'])
myexplode=np.array([.3,0,0,0])
plt.pie(x,labels=mylabels,explode=myexplode,shadow=True)
```

```python
#Matplotlib Pie Charts
#Colors
x=np.array([30,25,40,35])
mylabels=np.array(['task','home','school','sleep'])
myexplode=np.array([.3,0,0,0])
mycolors=np.array(["y", "hotpink", "m", "#4CAF50"])
plt.pie(x,labels=mylabels,explode=myexplode,shadow=True,colors=mycolors)
```

```python
#Matplotlib Pie Charts
#Legend
x=np.array([30,25,40,35])
mylabels=np.array(['task','home','school','sleep'])
myexplode=np.array([.3,0,0,0])
mycolors=np.array(["y", "hotpink", "m", "#4CAF50"])
plt.pie(x,labels=mylabels,explode=myexplode,shadow=True,colors=mycolors)
plt.legend()
```

```python
#Matplotlib Pie Charts
#Legend With Header
x=np.array([30,25,40,35])
mylabels=np.array(['task','home','school','sleep'])
myexplode=np.array([.3,0,0,0])
mycolors=np.array(["y", "hotpink", "m", "#4CAF50"])
plt.pie(x,labels=mylabels,explode=myexplode,shadow=True,colors=mycolors)
plt.legend(title='my day')
```
